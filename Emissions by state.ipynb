```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
file_path = r"C:\Users\user\Downloads\cleaned_emissions_data_v2.xlsx"
data_cleaned = pd.read_excel(file_path)

# Select the top 5 states based on 2022 emissions
top_states_2022 = data_cleaned.sort_values(by=2022, ascending=False).head(5)

# Plot the emissions trend for each of the top 5 states
plt.figure(figsize=(10, 6))
for state in top_states_2022["State"]:
    plt.plot(range(1970, 2023), top_states_2022[top_states_2022["State"] == state].iloc[0, 1:54], label=state)

# Format the plot
plt.title("Top 5 States with Highest CO Emissions (1970-2022)")
plt.xlabel("Year")
plt.ylabel("CO Emissions (Million Metric Tons)")
plt.legend(title="State")
plt.grid(True)
plt.show()
```

```python
# Bar chart for state-level emissions in 2022
plt.figure(figsize=(12, 8))
data_cleaned_sorted = data_cleaned.sort_values(by=2022, ascending=False)
plt.barh(data_cleaned_sorted["State"], data_cleaned_sorted[2022], color='skyblue')
plt.title("CO Emissions by State in 2022")
plt.xlabel("CO Emissions (Million Metric Tons)")
plt.ylabel("State")
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
# Convert the 'Absolute Change (2021-2022)' column to numeric, forcing errors to NaN and dropping them
data_cleaned['Absolute Change (2021-2022)'] = pd.to_numeric(data_cleaned['Absolute Change (2021-2022)'], errors='coerce')

# Remove rows where 'Absolute Change (2021-2022)' is NaN (not a number)
change_sorted = data_cleaned.dropna(subset=['Absolute Change (2021-2022)']).sort_values(by='Absolute Change (2021-2022)', ascending=False)

# Bar chart for absolute change in CO emissions (2021-2022)
plt.figure(figsize=(12, 8))
plt.barh(change_sorted["State"], change_sorted['Absolute Change (2021-2022)'], color='coral')
plt.title("Absolute Change in CO Emissions (2021-2022) by State")
plt.xlabel("Change in CO Emissions (Million Metric Tons)")
plt.ylabel("State")
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```

```python
import pandas as pd

# Load the cleaned dataset
file_path = r"C:\Users\user\Downloads\cleaned_emissions_data_v2.xlsx"  # Update this with your actual path
data_cleaned = pd.read_excel(file_path)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap for emissions data (1970-2022)
state_emissions = data_cleaned.set_index("State").loc[:, 1970:2022].transpose()
correlation_matrix = state_emissions.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of CO Emissions Across States (1970-2022)")
plt.show()
```

```python
# Track top 10 states in emissions over time
top_states = data_cleaned.sort_values(by=2022, ascending=False).head(10)

plt.figure(figsize=(10, 6))
for state in top_states["State"]:
    plt.plot(range(1970, 2023), top_states[top_states["State"] == state].iloc[0, 1:54], label=state)

plt.title("Top 10 States with Highest CO Emissions Over Time (1970-2022)")
plt.xlabel("Year")
plt.ylabel("CO Emissions (Million Metric Tons)")
plt.legend(title="State", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
